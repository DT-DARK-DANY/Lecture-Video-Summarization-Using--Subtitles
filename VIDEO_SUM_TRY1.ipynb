```python
def removeStopWords(text,lang):
  nltk.download('stopwords')
  stop_words = set(stopwords.words(lang))
  text = text.split()
  newText = ""
  for word in text:
    if word not in stop_words:
      newText += word+" "
  return newText.strip()

def applyFuncs(Data, textLabel):
  Data[textLabel] = Data[textLabel].apply(lambda x : x.lower())
  Data[textLabel] = Data[textLabel].apply(lambda x : re.sub("[^a-z0-9\s]","",x))
  Data[textLabel] = Data[textLabel].apply(lambda x : re.sub("\s+"," ",x))
  Data[textLabel] = Data[textLabel].apply(lambda x : removeStopWords(x,languages[result["language"]]))
  return Data
```

```python
from sentence_transformers import SentenceTransformer, util
from summarizer import Summarizer
```

```python
from transformers import BertTokenizer, EncoderDecoderModel
```

```python

tokenizer = BertTokenizer.from_pretrained("kairaamilanii/BERT2BERT-ExtractiveSummarizer")
tokenizer.bos_token = tokenizer.cls_token
tokenizer.eos_token = tokenizer.sep_token
model = EncoderDecoderModel.from_pretrained("kairaamilanii/BERT2BERT-ExtractiveSummarizer")
```

```python

embedder = SentenceTransformer('all-MiniLM-L6-v2')
```

```python

def summarize_text(text):
    print(len(text))
    input_ids = tokenizer.encode(text, return_tensors='pt')
    summary_ids = model.generate(input_ids,
            min_length=20,
            max_length=80,
            num_beams=10,
            repetition_penalty=2.5,
            length_penalty=1.0,
            early_stopping=True,
            no_repeat_ngram_size=2,
            use_cache=True,
            do_sample = True,
            temperature = 0.8,
            top_k = 50,
            top_p = 0.95)

    summary_text = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
    return summary_text
```

```python


def dynamic_topic_segmentation(data, similarity_threshold=0.7):
    embeddings = embedder.encode(data['text'].tolist(), convert_to_tensor=True)

    topic_segments = []
    current_segment = []
    start_time = data['start'][0]
    end_time = data['end'][0]

    for i in range(1, len(embeddings)):
        similarity = util.pytorch_cos_sim(embeddings[i-1], embeddings[i]).item()

        if similarity < similarity_threshold:
            segment_text = " ".join(current_segment)
            summary = summarize_text(segment_text)

            topic_segments.append({
                'start_time': start_time,
                'end_time': end_time,
                'summary': summary
            })

            current_segment = []
            start_time = data['start'][i]
            end_time = data['end'][i]

        current_segment.append(data['text'][i])

    if current_segment:
        segment_text = " ".join(current_segment)
        summary = summarize_text(segment_text)
        end_time = data['end'].iloc[-1]
        topic_segments.append({
            'start_time': start_time,
            'end_time': end_time,
            'summary': summary
        })

    return topic_segments

dynamic_summaries = dynamic_topic_segmentation(DATA, similarity_threshold=0.7)
for segment in dynamic_summaries:
    print(f"({segment['start_time']}s - {segment['end_time']}s): {segment['summary']}")
```
