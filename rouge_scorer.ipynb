```python
def get_summ_real_Data(DATA,all_sum):
  all_text = DATA["text"].tolist() + all_sum
  vectorizer = TfidfVectorizer().fit_transform(all_text)
  vectors = vectorizer.toarray()
  similarity_matrix = cosine_similarity(vectors[:len(DATA)], vectors[len(DATA):])
  df=DATA.copy()
  threshold = 0.4
  df["best_match"] = ""
  df["similarity_score"] = 0.0
  for j, subtitle_vector in enumerate(similarity_matrix):
    best_match_idx = subtitle_vector.argmax()
    best_score = subtitle_vector[best_match_idx]
    if best_score > threshold:
        df.at[j, "best_match"] = all_sum[best_match_idx]
        df.at[j, "similarity_score"] = best_score
  matched_df = df[df["similarity_score"] > threshold]
  return matched_df
```

```python
scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)

def calculate_rouge_scores(reference, generated):
    scores = scorer.score(reference, generated)
    return {k: v.fmeasure for k, v in scores.items()}

ground_truth_Definition_of_Information_p = " ".join(Definition_of_Information_p['text'].tolist())
ground_truth_Recoverable_schdeules_p = " ".join(Recoverable_schdeules_p['text'].tolist())
ground_truth_Introduction_to_OS_p = " ".join(Introduction_to_OS_p['text'].tolist())

ground_truth_Introduction_to_Machine_p = " ".join(Introduction_to_Machine_p['text'].tolist())
#######################################
print("#######################################")
print("_TRY2__removing_stop_words")
Definition_of_Information_TRY2__removing_stop_words_T = " ".join(Definition_of_Information_TRY2__removing_stop_words['text'].tolist())

print("Definition_of_Information")
print(calculate_rouge_scores(ground_truth_Definition_of_Information_p, Definition_of_Information_TRY2__removing_stop_words_T))

Recoverable_schdeules_TRY2__removing_stop_words_T = " ".join(Recoverable_schdeules_TRY2__removing_stop_words['text'].tolist())

print("Recoverable_schdeules")
print(calculate_rouge_scores(ground_truth_Recoverable_schdeules_p, Recoverable_schdeules_TRY2__removing_stop_words_T))

Introduction_to_OS_TRY2__removing_stop_words_T = " ".join(Introduction_to_OS_TRY2__removing_stop_words['text'].tolist())

print("Introduction_to_OS")
print(calculate_rouge_scores(ground_truth_Introduction_to_OS_p, Introduction_to_OS_TRY2__removing_stop_words_T))

Introduction_to_Machine_TRY2__removing_stop_words_T = " ".join(Introduction_to_Machine_TRY2__removing_stop_words['text'].tolist())

print("ground_truth_Introduction_to_Machine")
print(calculate_rouge_scores(ground_truth_Introduction_to_Machine_p, Introduction_to_Machine_TRY2__removing_stop_words_T))
#######################################
print("#######################################")
print("_TRY2__without_removing_stop_words")

Definition_of_Information_TRY2__without_removing_stop_words_T = " ".join(Definition_of_Information_TRY2__without_removing_stop_words['text'].tolist())
print("Definition_of_Information")
print(calculate_rouge_scores(ground_truth_Definition_of_Information_p, Definition_of_Information_TRY2__without_removing_stop_words_T))

Recoverable_schdeules_TRY2__without_removing_stop_words_T = " ".join(Recoverable_schdeules_TRY2__without_removing_stop_words['text'].tolist())

print("Recoverable_schdeules")
print(calculate_rouge_scores(ground_truth_Recoverable_schdeules_p, Recoverable_schdeules_TRY2__without_removing_stop_words_T))

Introduction_to_OS_TRY2__without_removing_stop_words_T = " ".join(Introduction_to_OS_TRY2__without_removing_stop_words['text'].tolist())

print("Introduction_to_OS")
print(calculate_rouge_scores(ground_truth_Introduction_to_OS_p, Introduction_to_OS_TRY2__without_removing_stop_words_T))

Introduction_to_Machine_TRY2__without_removing_stop_words_T = " ".join(Introduction_to_Machine_TRY2__without_removing_stop_words['text'].tolist())

print("ground_truth_Introduction_to_Machine")
print(calculate_rouge_scores(ground_truth_Introduction_to_Machine_p, Introduction_to_Machine_TRY2__without_removing_stop_words_T))
#######################################
print("#######################################")
print("_TRY3")

Definition_of_Information_TRY3_T = " ".join(Definition_of_Information_TRY3['text'].tolist())
print("Definition_of_Information")
print(calculate_rouge_scores(ground_truth_Definition_of_Information_p, Definition_of_Information_TRY3_T))

Recoverable_schdeules_TRY3_T = " ".join(Recoverable_schdeules_TRY3['text'].tolist())
print("Recoverable_schdeules")
print(calculate_rouge_scores(ground_truth_Recoverable_schdeules_p, Recoverable_schdeules_TRY3_T))

Introduction_to_OS_TRY3_T = " ".join(Introduction_to_OS_TRY3['text'].tolist())
print("Introduction_to_OS")
print(calculate_rouge_scores(ground_truth_Introduction_to_OS_p, Introduction_to_OS_TRY3_T))

Introduction_to_Machine_TRY3_T = " ".join(Introduction_to_Machine_TRY3['text'].tolist())

print("ground_truth_Introduction_to_Machine")
print(calculate_rouge_scores(ground_truth_Introduction_to_Machine_p, Introduction_to_Machine_TRY3_T))
#######################################
print("#######################################")
print("_TRY4_LLM")

Definition_of_Information_LLM_T = " ".join(Definition_of_Information_LLM['text'].tolist())
print("Definition_of_Information")
print(calculate_rouge_scores(ground_truth_Definition_of_Information_p, Definition_of_Information_LLM_T))

Recoverable_schdeules_LLM_T = " ".join(Recoverable_schdeules_LLM['text'].tolist())
print("Recoverable_schdeules")
print(calculate_rouge_scores(ground_truth_Recoverable_schdeules_p, Recoverable_schdeules_LLM_T))

Introduction_to_OS_LLM_T = " ".join(Introduction_to_OS_LLM['text'].tolist())
print("Introduction_to_OS")
print(calculate_rouge_scores(ground_truth_Introduction_to_OS_p, Introduction_to_OS_LLM_T))

Introduction_to_Machine_LLM_T = " ".join(Introduction_to_Machine_LLM['text'].tolist())
print("ground_truth_Introduction_to_Machine")
print(calculate_rouge_scores(ground_truth_Introduction_to_Machine_p, Introduction_to_Machine_LLM_T))

print("#######################################")

```
